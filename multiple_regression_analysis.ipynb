
# Multiple Linear Regression – Multi-Channel Marketing Analysis

## Project Overview
This project evaluates how TV, Radio, and Social Media marketing channels influence Sales using Multiple Linear Regression.


```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.stats.diagnostic import het_breuschpagan

# Load dataset
df = pd.read_csv('marketing_sales_data.csv')

# Preview dataset
df.head()
```

```

# Dataset overview
print(df.info())

# Missing values
print(df.isnull().sum())

# Drop missing values
df = df.dropna()

# Summary statistics
df.describe()
```


# Exploratory Data Analysis


```

# Correlation matrix
corr = df.corr(numeric_only=True)

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='Blues')
plt.title('Correlation Matrix')
plt.show()
```

```

# Pairplot
sns.pairplot(df)
plt.show()
```


# Multicollinearity Check Using VIF


```

# Independent variables
X = df[['TV', 'Radio', 'Social_Media']]

# Add constant
X_const = sm.add_constant(X)

# Calculate VIF
vif_data = pd.DataFrame()
vif_data['Variable'] = X_const.columns
vif_data['VIF'] = [variance_inflation_factor(X_const.values, i) for i in range(X_const.shape[1])]

print(vif_data)
```


# Build Multiple Linear Regression Model


```

# Dependent variable
y = df['Sales']

# Add constant
X = sm.add_constant(X)

# Fit model
model = sm.OLS(y, X).fit()

# Model summary
print(model.summary())
```


# Diagnostic Plots


```

# Predictions and residuals
predictions = model.predict(X)
residuals = y - predictions

# Residual plot
plt.figure(figsize=(6,4))
sns.scatterplot(x=predictions, y=residuals)
plt.axhline(0, linestyle='--')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs Fitted')
plt.show()

# Residual histogram
plt.figure(figsize=(6,4))
plt.hist(residuals, bins=20)
plt.title('Residual Distribution')
plt.show()

# QQ plot
sm.qqplot(residuals, line='45')
plt.title('QQ Plot')
plt.show()
```

```

# Breusch-Pagan test
bp_test = het_breuschpagan(residuals, X)

labels = ['LM Statistic', 'LM-Test p-value', 'F-Statistic', 'F-Test p-value']

print(dict(zip(labels, bp_test)))
```


# Interpretation

## Adjusted R-squared
Shows how well the independent variables explain variation in Sales.

## Coefficients
Each coefficient measures the expected change in Sales from a one-unit increase in the predictor while holding other predictors constant.

## p-values
Variables with p-values below 0.05 are statistically significant.

## Business Recommendation
Invest more in channels with statistically significant positive coefficients to maximize ROI.
